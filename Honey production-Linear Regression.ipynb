```python
# group the data by states and using sum method to get the total honey production value in decending order
us_state=hp[['state','totalprod','yieldpercol']].groupby('state').sum()
us_state.reset_index(level=0,inplace=True)
us_state.sort_values(by='totalprod',ascending=False,inplace=True)
us_state.head()
```

```python
#creating a bar chart to visualize the toatal honey production
plt.figure(figsize=(20,7))
sns.barplot(x=us_state['state'],y=us_state['totalprod'])
plt.title('statewise total honey production',fontsize=20)
plt.xlabel('states',fontsize=20)
plt.ylabel('total production of honey in USA',fontsize=20)
```

```python
#creating a table to find maximum production value from the state
us_state_max= hp[['state','totalprod']].groupby('state').max()
us_state_max.reset_index(level=0,inplace=True)
us_state_max.columns=['state','Max prod']
us_state_max.head()
```

```python
us_state_min=hp[['state','totalprod']].groupby('state').min()
us_state_min.reset_index(level=0,inplace=True)
us_state_min.columns=['state','Min prod']
us_state_min.head()
```

```python
#merginf max and min to find range
st_range=pd.merge(us_state_max,us_state_min,how='inner',on='state')
st_range.head()
```

```python
st_range['per_change']=((st_range['Max prod']-st_range['Min prod'])/st_range['Max prod'])*100
st_range.sort_values(by='per_change',ascending=False,inplace=True)
st_range.head()
```

```python
#Xtrain: all 80% column except totalprod
#Xtest: all 20% column except totalprod
#ytrain: 80% values of totalprod
#ytest: 20% values of totalprod
#linear regression is for continuous values like age salary price numeric value
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
"""test_size = 0.20 means to leave 20% data unseen to the model so that we can check accuracy with this 
we will have actual and predicted values for this 20% data and accordingly will calculate accuracy
random_state: selecting values randomly so that our model is not biased. ( selecting 1st row after that 9th and so on) repeating this till 80% of data is not selected"""
x_train,x_test,y_train,y_test=train_test_split(hp.drop(['totalprod','state'],axis=1),
                                               hp['totalprod'],test_size=0.20,
                                              random_state=8)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
accuracy=model.score(x_test,y_test)
print(accuracy*100,'%')
```
